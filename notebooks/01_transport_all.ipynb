```python
#Load libraries
import pandas as pd
from configobj import ConfigObj
import os
import sys
import math
import sqlalchemy
import numpy as np

module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)
    
from functions.transport_by_tower_owner import transport_by_tower_owner
from functions.utils import print_full, elevation_google, database
```

```python
## define paths and variables

config_path = "/home/jovyan/shared/rural_planner_refactoring/config_files_test/config_co"

parser = ConfigObj(config_path)

sql_path =  parser['sql_path']
country_folder = parser['country_folder']
schema = parser['transport_by_tower_params']['schema']


output_table = parser['transport_by_tower_params']['output_table']
temporary_table = parser['transport_by_tower_params']['temporary_table']
table_infrastructure = parser['transport_by_tower_params']['table_infrastructure']

sources_omit = parser['transport_by_tower_params']['sources_omit']
owners_tx = dict(parser['transport_by_tower_params']['owners_tx'])

third_party_owners = parser['transport_by_tower_params']['third_party_owners']
regional_owners = parser['transport_by_tower_params']['regional_owners']
tef_alias = (owners_tx.keys() - (regional_owners + third_party_owners)).pop()
```

```python
## Loop through owners dictionary and get best transport towers by owner
query_path = sql_path + '/' + country_folder + '/' + 'transport_by_tower_load_towers.sql'
with open(query_path) as file, database(parser) as db:
    query = file.read()
    query_formatted = query.format(schema = schema, table = table_infrastructure, sources_omit = (','.join('\'{0}\''.format(s) for s in sources_omit)))
    df_output = pd.read_sql_query(query_formatted, db)
        
query_path = sql_path + '/' + country_folder + '/' + 'transport_by_tower_inputs.sql'
query_path_all = sql_path + '/' + country_folder + '/' + 'transport_by_tower_all_inputs.sql'

for i in owners_tx.keys():
    owners = owners_tx[i]['owners']
    owner_name = owners_tx[i]['owner_name']
    df_towers_owner = transport_by_tower_owner(parser, owners, owner_name, query_path, query_path_all, column_name = 'tower_id')
    df_output = df_output.merge(df_towers_owner, on='tower_id', how='left')
    del owners, owner_name, df_towers_owner
```

```python
## Define third party colnames
third_party_transport_colnames = [(owner_name + '_transport_id') for owner_name in (third_party_owners)]
third_party_distance_colnames = [('distance_' + owner_name + '_transport_m') for owner_name in (third_party_owners)]
third_party_line_of_sight_colnames = [('line_of_sight_' + owner_name) for owner_name in (third_party_owners)]
third_party_additional_height_1_colnames = [('additional_height_tower_1_' + owner_name + '_m') for owner_name in (third_party_owners)]
third_party_additional_height_2_colnames = [('additional_height_tower_2_' + owner_name + '_m') for owner_name in (third_party_owners)]

## Define regional colnames
regional_transport_colnames = [(owner_name + '_transport_id') for owner_name in (regional_owners)]
regional_distance_colnames = [('distance_' + owner_name + '_transport_m') for owner_name in (regional_owners)]
regional_line_of_sight_colnames = [('line_of_sight_' + owner_name) for owner_name in (regional_owners)]
regional_additional_height_1_colnames = [('additional_height_tower_1_' + owner_name + '_m') for owner_name in (regional_owners)]
regional_additional_height_2_colnames = [('additional_height_tower_2_' + owner_name + '_m') for owner_name in (regional_owners)]
```

```python
# Initialize columns
df_output['regional_transport_id'] = math.nan
df_output['line_of_sight_regional_transport'] = False
df_output['distance_regional_transport_m'] = math.nan
df_output['additional_height_tower_1_regional_m'] = math.nan
df_output['additional_height_tower_2_regional_m'] = math.nan

df_output['third_party_transport_id'] = math.nan
df_output['line_of_sight_third_party_transport'] = False
df_output['distance_third_party_transport_m'] = math.nan
df_output['additional_height_tower_1_third_party_m'] = math.nan
df_output['additional_height_tower_2_third_party_m'] = math.nan

## Assign first non-null owner according to input priorization 

if (len(regional_transport_colnames)>0):
    df_output['regional_transport_id'] = df_output[[s for s in regional_transport_colnames]].bfill(axis=1).iloc[:, 0].fillna(np.nan)
    df_output['line_of_sight_regional_transport'] = df_output[[s for s in regional_line_of_sight_colnames]].bfill(axis=1).iloc[:, 0].fillna(np.nan)
    df_output['distance_regional_transport_m'] = df_output[[s for s in regional_distance_colnames]].bfill(axis=1).iloc[:, 0].fillna(np.nan)
    df_output['additional_height_tower_1_regional_m'] = df_output[[s for s in regional_additional_height_1_colnames]].bfill(axis=1).iloc[:, 0].fillna(np.nan)
    df_output['additional_height_tower_2_regional_m'] = df_output[[s for s in regional_additional_height_2_colnames]].bfill(axis=1).iloc[:, 0].fillna(np.nan)

if (len(third_party_transport_colnames)>0):
    df_output['third_party_transport_id'] = df_output[[s for s in third_party_transport_colnames]].bfill(axis=1).iloc[:, 0].fillna(np.nan)
    df_output['line_of_sight_third_party_transport'] = df_output[[s for s in third_party_line_of_sight_colnames]].bfill(axis=1).iloc[:, 0].fillna(np.nan)
    df_output['distance_third_party_transport_m'] = df_output[[s for s in third_party_distance_colnames]].bfill(axis=1).iloc[:, 0].fillna(np.nan)
    df_output['additional_height_tower_1_third_party_m'] = df_output[[s for s in third_party_additional_height_1_colnames]].bfill(axis=1).iloc[:, 0].fillna(np.nan)
    df_output['additional_height_tower_2_third_party_m'] = df_output[[s for s in third_party_additional_height_2_colnames]].bfill(axis=1).iloc[:, 0].fillna(np.nan)
```

```python
#Export output

query_path = sql_path + '/' + 'transport_by_tower_add_geoms.sql'

with open(query_path) as file, database(parser) as db:    
    df_output.to_sql(output_table, con=db, if_exists = 'replace', schema = schema, index = False, 
                dtype = {'tower_id': sqlalchemy.types.Integer(),
                         tef_alias + '_transport_id': sqlalchemy.types.Integer(), 
                         'distance_' + tef_alias + '_transport_m': sqlalchemy.types.Integer(),
                         'line_of_sight_' + tef_alias: sqlalchemy.types.Boolean(), 
                         'additional_height_tower_1_' + tef_alias + '_m': sqlalchemy.types.Integer(),
                         'additional_height_tower_2_' + tef_alias + '_m': sqlalchemy.types.Integer(),
                         'regional_transport_id': sqlalchemy.types.Integer(), 
                         'distance_regional_transport_m': sqlalchemy.types.Integer(),
                         'line_of_sight_regional': sqlalchemy.types.Boolean(), 
                         'additional_height_tower_1_regional_m': sqlalchemy.types.Integer(),
                         'additional_height_tower_2_regional_m': sqlalchemy.types.Integer(),
                         'third_party_transport_id': sqlalchemy.types.Integer(),
                         'line_of_sight_third_party_transport': sqlalchemy.types.Boolean(), 
                         'distance_third_party_transport_m': sqlalchemy.types.Integer(),
                         'additional_height_tower_1_third_party_m': sqlalchemy.types.Integer(),
                         'additional_height_tower_2_third_party_m': sqlalchemy.types.Integer(), 
                         (s for s in regional_transport_colnames): sqlalchemy.types.Integer(),
                         (s for s in regional_distance_colnames): sqlalchemy.types.Integer(),
                         (s for s in regional_line_of_sight_colnames): sqlalchemy.types.Boolean(),
                         (s for s in regional_additional_height_1_colnames): sqlalchemy.types.Integer(),
                         (s for s in regional_additional_height_2_colnames): sqlalchemy.types.Integer(), 
                         (s for s in third_party_transport_colnames): sqlalchemy.types.Integer(),
                         (s for s in third_party_distance_colnames): sqlalchemy.types.Integer(),
                         (s for s in third_party_line_of_sight_colnames): sqlalchemy.types.Boolean(),
                         (s for s in third_party_additional_height_1_colnames): sqlalchemy.types.Integer(),
                         (s for s in third_party_additional_height_2_colnames): sqlalchemy.types.Integer()                         
                        })
    query = file.read()
    query_formatted = query.format(schema = schema,
                                   output_table = output_table,
                                   temporary_table = temporary_table,
                                   infrastructure_table = table_infrastructure,
                                   tef_alias = tef_alias)
    db.execute(query_formatted)
```
